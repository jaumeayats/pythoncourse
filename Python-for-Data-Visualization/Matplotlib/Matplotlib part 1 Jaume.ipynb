```python
x = np.linspace(0,5,11)
y = x**2
```

```python
plt.plot(x,y,'r.')
```

```python
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r-')

plt.subplot(1,2,2)
plt.plot(y,x,'b.')
```

```python
#OO when doing object oriented you work more through methods
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('INSERT 1')
axes2.plot(y,x)
axes2.set_title('INSERT 2')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.,0.,1,1])

axes1.plot(x,y)
```
